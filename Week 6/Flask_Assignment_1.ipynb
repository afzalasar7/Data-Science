# Q1. What is Flask Framework? What are the advantages of Flask Framework?
A1. Flask is a micro web framework for Python that provides the necessary tools and libraries to build web applications. It is lightweight, flexible, and designed to be easy to use. Flask follows the model-view-controller (MVC) architectural pattern and provides a simple and intuitive interface to handle HTTP requests and responses.

Advantages of Flask Framework include:

- Lightweight and minimalistic: Flask is a minimalistic framework that does not impose any specific tools or libraries. It allows developers to choose and integrate only the components they need, resulting in a lightweight application with reduced overhead.

- Easy to get started: Flask has a simple and intuitive API, making it easy for developers, especially beginners, to get started with web development. Its documentation is extensive and provides clear examples and explanations.

- Flexible and extensible: Flask allows developers to customize and extend the framework according to their project requirements. It supports various extensions and libraries that can be easily integrated, providing additional functionality and flexibility.

- Built-in development server: Flask comes with a built-in development server, allowing developers to quickly test and run their applications without the need for additional setup. This makes the development process faster and more efficient.

- Large community and active ecosystem: Flask has a large and active community of developers who contribute to its development and provide support. The ecosystem around Flask includes numerous third-party extensions, libraries, and resources, making it easier to find solutions and integrations for specific needs.

# Q3. What is App routing in Flask? Why do we use app routes?
A3. App routing in Flask refers to the process of mapping URLs to specific functions in a Flask application. It allows defining different routes or URLs that users can visit and specifying the corresponding function or view that should be executed when a particular URL is accessed.

We use app routes in Flask for the following reasons:

- URL handling: App routes define the URLs or routes that users can access in a Flask application. By associating each route with a specific function or view, Flask knows what code to execute when a particular URL is requested.

- Resource organization: App routes help organize the application's resources and define the structure of the web application. They provide a clear and logical way to map different URLs to specific functionalities or views.

- Request handling: App routes allow developers to handle different types of HTTP requests, such as GET, POST, PUT, DELETE, etc., for specific URLs. This enables building interactive web applications that respond differently based on the type of request received.

- Dynamic routing: Flask supports dynamic routing by allowing the use of variables or placeholders in routes. This allows capturing dynamic values from URLs and passing them as parameters to the associated view function.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
A5. The `url_for()` function is used in Flask for URL building. It generates a URL for a specific function or view based on its name and any associated arguments.

Here's an example code to demonstrate the working of the `url_for()` function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/user/<username>')
def profile(username):
    return f'Profile Page of {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))  # Output: /
        print(url_for('about'))  # Output: /about
        print(url_for('profile', username='John'))  # Output: /user/John
```

In this code, the `url_for()` function is used to generate URLs for the `home`, `about`, and `profile` routes. The function takes the name of the route's view function as the argument and, optionally, any additional arguments required by the route. It returns the URL corresponding to the provided route and arguments.

The output of the `url_for()` function in the example code would be:

```
/
/about
/user/John
```
