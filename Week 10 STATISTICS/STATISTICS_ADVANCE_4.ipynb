## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

**Answer:**
A **t-test** and a **z-test** are both statistical hypothesis tests used to make inferences about population parameters based on sample data. However, they are applied in different situations due to variations in sample size and our knowledge about population parameters.

**T-test:** A t-test is used when the sample size is small (typically less than 30) and the population standard deviation is unknown. For example, if you have a sample of 20 students and want to test whether their average test scores are significantly different from a known mean (population mean), you would use a t-test.

**Z-test:** A z-test is used when the sample size is relatively large (typically greater than 30) and/or the population standard deviation is known. For instance, if you have a sample of 1000 products and want to determine if their average weight is significantly different from a known population mean weight, you would use a z-test.

## Q2: Differentiate between one-tailed and two-tailed tests.

**Answer:**
In hypothesis testing, **one-tailed tests** and **two-tailed tests** refer to the directionality of the alternative hypothesis and the corresponding critical regions.

- **One-Tailed Test:** Also known as a directional test, it focuses on deviations in one specific direction from the null hypothesis. For example, if you're testing a new drug's effectiveness and you expect it to improve outcomes, you'd use a one-tailed test. The critical region is only on one side of the distribution.

- **Two-Tailed Test:** Also known as a non-directional test, it considers deviations in both directions from the null hypothesis. This is useful when you're interested in any significant deviation, not just in one specific direction. For instance, if you're testing whether a coin is fair, you'd use a two-tailed test. The critical region is divided between both sides of the distribution.

## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

**Answer:**
**Type 1 Error (False Positive):** This occurs when you reject the null hypothesis when it is actually true. In other words, you conclude there is an effect when there isn't. For example, if a medical test incorrectly diagnoses a healthy person as having a disease, that's a type 1 error.

**Type 2 Error (False Negative):** This happens when you fail to reject the null hypothesis when it is actually false. It means you miss detecting a real effect. For instance, if a medical test fails to diagnose a person with a disease when they actually have it, that's a type 2 error.

Example Scenarios:
1. **Type 1 Error:** A jury mistakenly convicts an innocent person due to insufficient evidence.
2. **Type 2 Error:** A company fails to recognize a defective product in quality control testing, leading to customer dissatisfaction.

## Q4: Explain Bayes's theorem with an example.

**Answer:**
Bayes's theorem is a mathematical formula that describes how to update the probability for a hypothesis as more evidence or information becomes available.

Formula: P(A|B) = (P(B|A) * P(A)) / P(B)

Example: Consider a medical test to detect a rare disease. Let A be the event that a person has the disease, and B be the event that the test result is positive.

- P(A): Prior probability of having the disease.
- P(B|A): Probability of a positive test given that the person has the disease.
- P(B): Overall probability of a positive test.

Using Bayes's theorem, you can calculate the probability that a person has the disease given a positive test result.

## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

**Answer:**
A **confidence interval** is a range of values that is likely to contain the true population parameter (e.g., mean or proportion) with a certain level of confidence. It quantifies the uncertainty around the sample estimate.

Calculation:
Confidence Interval = Sample Mean ± (Critical Value) * (Standard Error)

Example: Suppose you have a sample of 100 students and want to estimate the population mean height at a 95% confidence level. If the sample mean height is 165 cm and the standard deviation is 10 cm, the critical value for a 95% confidence level is approximately 1.96. The standard error would be 10 / √100 = 1. Plug these values into the formula to calculate the confidence interval.

Solution using Bayes's Theorem:
Let A be the event that the product is A, and D be the event that the product is defective.

- P(D) = P(A) * P(D|A) + P(B) * P(D|B)

Now use the formula:
P(A|D) = (0.05 * 0.5) / 0.075 = 0.3333

## Q

8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

**Answer:**
The **margin of error** (MOE) in a confidence interval is the range around the sample estimate (mean, proportion, etc.) that accounts for the potential variability between the sample and the population. It's a measure of the uncertainty in our estimate.

Formula for MOE:
MOE = Critical Value * Standard Error

Sample size directly affects the margin of error. As the sample size increases, the standard error decreases, resulting in a smaller margin of error. This is because larger samples provide more accurate estimates of the population parameter.

Example: Consider a survey to estimate the average age of citizens in a city. With a smaller sample size (say, 50), the margin of error might be ±3 years. With a larger sample size (say, 500), the margin of error could be reduced to ±1 year. This means that the larger sample size provides a more precise estimate with a smaller margin of error.

## Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

**Answer:**
The **z-score** measures how many standard deviations a data point is away from the population mean. It's calculated using the formula:

z = (x - μ) / σ

Given:
- Data Point (x) = 75
- Population Mean (μ) = 70
- Population Standard Deviation (σ) = 5

Calculating the z-score:
z = (75 - 70) / 5 = 1

Interpretation: The z-score of 1 indicates that the data point (75) is 1 standard deviation above the population mean (70). This tells us that the data point is relatively higher than the average value in the population.

## Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

**Answer:**
Null Hypothesis (H0): The new weight loss drug is not significantly effective.
Alternative Hypothesis (Ha): The new weight loss drug is significantly effective.

Given:
Sample Mean (x̄) = 6 pounds
Sample Standard Deviation (s) = 2.5 pounds
Sample Size (n) = 50
Confidence Level = 95%
Degrees of Freedom (df) = n - 1 = 50 - 1 = 49

Critical Value: For a 95% confidence level and two-tailed test, the critical t-value is approximately ±2.0096 (look up in t-distribution table or calculator).

Calculating the t-statistic:
t = (x̄ - μ) / (s / √n) = (6 - 0) / (2.5 / √50) = 8.49

Since the calculated t-statistic (8.49) is greater than the critical t-value (±2.0096), we can reject the null hypothesis.

Conclusion: At a 95% confidence level, we have evidence to conclude that the new weight loss drug is significantly effective.

## Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

**Answer:**
To calculate the confidence interval for a proportion (p), use the formula:

Confidence Interval = p ± (Critical Value) * √((p * (1 - p)) / n)

Given:
Sample Proportion (p) = 0.65 (65%)
Sample Size (n) = 500
Confidence Level = 95%

For a 95% confidence level and two-tailed test, the critical z-value is approximately ±1.96 (look up in standard normal distribution table).

Calculating the confidence interval:
Confidence Interval = 0.65 ± 1.96 * √((0.65 * (1 - 0.65)) / 500)

Interpretation: This means we are 95% confident that the true proportion of people satisfied with their job lies within the calculated interval.

 = ?

Calculating the t-statistic:
t = ((x̄A - x̄B) - 0) / √((sA^2 / nA) + (sB^2 / nB))

## Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.

**Answer:**

Confidence Interval = Sample Mean ± (Critical Value) * (Sample Standard Deviation / √n)

Given:
Population Mean (μ) = 60
Population Standard Deviation (σ) = 8
Sample Size (n) = 50
Confidence Level = 90%

Assuming a large enough sample size (approx. n > 30), the critical value for a 90% confidence level is approximately 1.645 (look up in standard normal distribution table).

Calculating the confidence interval:
Confidence Interval = 65 ± 1.645 * (8 / √50)

Interpretation: This means we are 90% confident that the true population mean lies within the calculated interval.
