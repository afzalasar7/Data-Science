# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data, plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```



A2: A scatter plot is a type of plot that displays individual data points as markers, where each marker represents two different variables. It is used to visualize the relationship between two numerical variables and helps identify patterns, correlations, or clusters in the data.

In the given code, we generate random data for `x` and `y` using numpy and then plot a scatter plot using Matplotlib. The `plt.scatter()` function is used to create the scatter plot, and `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` functions are used to add a title, x-axis label, and y-axis label to the plot, respectively.


# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.


```python

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Data for four line plots
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```



A3: The `subplot()` function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to display multiple plots in a grid layout, making it easier to compare and analyze different visualizations side by side.

In the given code, we create four line plots using the `plt.plot()` function and then use the `plt.subplot()` function to arrange these plots in a 2x2 grid layout. Each subplot displays one of the four line plots with different data. The `plt.title()` function is used to add titles to each subplot, and `plt.tight_layout()` ensures that the subplots do not overlap when displayed.


# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure()
plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal bar plot
plt.figure()
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```



A4: A bar plot is a type of plot that represents categorical data using rectangular bars, where the length of each bar is proportional to the value of the corresponding category. It is used to compare and display discrete data, such as the performance of different categories or groups.

Bar plots are useful for visualizing data that can be divided into distinct categories or groups and are commonly used to show comparisons between different data points.

In the given code, we create a bar plot using the `plt.bar()` function to visualize the profit of different companies. We also create a horizontal bar plot using the `plt.barh()` function, which displays the same data but with the companies on the y-axis and the profit on the x-axis.


# Q5: What is a box plot? Why is it used? Using the following data, plot a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.show()
```

A5: A box plot (box-and-whisker plot) is a graphical representation of the distribution of numerical data. It displays the median, quartiles, and potential outliers of the data. The box represents the interquartile range (IQR), which spans the 25th to 75th percentiles of the data. The "whiskers" extend from the edges of the box to show the minimum and maximum values within a certain range (usually 1.5 times the IQR).

Box plots are used to identify the spread, skewness, and potential outliers in the data. They provide a compact way to visualize the distribution and central tendency of the data, making it easy to compare multiple datasets.

In the given code, we create a box plot using the `plt.boxplot()` function to visualize the distribution of data in `box1` and `box2`. The `plt.title()`, `plt.xticks()`, `plt.xlabel()`, and `plt.ylabel()` functions are used to add a title, modify x-axis labels, and add axis labels to the plot.
