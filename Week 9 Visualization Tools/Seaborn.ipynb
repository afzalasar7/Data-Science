# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans 1: Seaborn is a Python data visualization library based on Matplotlib. It provides a variety of plots to effectively visualize and analyze data. Here are five plots that we can create using Seaborn and their uses:

1. Scatter Plot:
   - Use: To visualize the relationship between two continuous variables. It helps identify patterns, clusters, and correlations in the data.

2. Line Plot:
   - Use: To display the trend or changes in a continuous variable over time or any other continuous dimension. It is commonly used to visualize time-series data.

3. Bar Plot:
   - Use: To compare categorical data across different categories. It is suitable for displaying the distribution of data or comparing the values of different groups.

4. Box Plot:
   - Use: To show the distribution of numerical data and identify potential outliers. It displays the median, quartiles, and potential range of the data.

5. Histogram:
   - Use: To represent the distribution of a continuous variable by dividing it into bins and counting the number of occurrences in each bin.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x="timepoint" and y="signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("Line Plot of Signal for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x="pclass", y="age" and y="fare".

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for 'age' and 'fare' based on 'pclass'
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram for 'price' with 'cut' as hue
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True)
plt.title("Histogram of Price with Cut as Hue")
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend(title="Cut")
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with 'species' as hue
sns.pairplot(data=iris_data, hue="species")
plt.title("Pair Plot of Iris Dataset with Species as Hue")
plt.show()
```


# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Reshape the data to create a pivot table
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='viridis', annot=True, fmt='d', linewidths=0.5)
plt.title("Heatmap of Flights Dataset")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
