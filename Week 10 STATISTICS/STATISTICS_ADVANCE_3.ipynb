# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
**Estimation Statistics** involves using sample data to estimate population parameters. It provides insight into the characteristics of the population based on limited sample information.

- **Point Estimate:** A single value that estimates a population parameter. For example, using the sample mean to estimate the population mean.

- **Interval Estimate (Confidence Interval):** A range of values within which the true population parameter is likely to fall. It quantifies the uncertainty associated with the estimation.

# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.


```python
import scipy.stats as stats
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level):
    degrees_of_freedom = sample_size - 1
    critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)
    margin_of_error = critical_value * (sample_std_dev / math.sqrt(sample_size))
    confidence_interval_lower = sample_mean - margin_of_error
    confidence_interval_upper = sample_mean + margin_of_error

    return sample_mean, (confidence_interval_lower, confidence_interval_upper)

# Given values
sample_mean = 50
sample_std_dev = 8
sample_size = 100
confidence_level = 0.95

estimated_mean, confidence_interval = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)

print(f"Estimated population mean: {estimated_mean}")
print(f"Confidence Interval: [{confidence_interval[0]:.4f}, {confidence_interval[1]:.4f}]")
```

# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
**Hypothesis testing** is a statistical method used to make decisions about population parameters based on sample data. It involves comparing a hypothesis (null hypothesis) with observed data to determine if there's enough evidence to reject the hypothesis.

Importance:
1. It allows us to make data-driven decisions in uncertain situations.
2. It helps in assessing the validity of assumptions and claims.
3. It's fundamental in scientific research, quality control, and decision-making.

# Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.
**Null Hypothesis (H0):** The average weight of male college students is not greater than the average weight of female college students.

**Alternative Hypothesis (Ha):** The average weight of male college students is greater than the average weight of female college students.


```python
def compare_weight_hypothesis(hypothesis):
    male_mean_weight = 170  # Example average weight of male students
    female_mean_weight = 150  # Example average weight of female students

    if hypothesis == "greater":
        return male_mean_weight > female_mean_weight
    elif hypothesis == "less":
        return male_mean_weight < female_mean_weight
    elif hypothesis == "not_equal":
        return male_mean_weight != female_mean_weight
    else:
        return "Invalid hypothesis"

# Example usage
hypothesis = "greater"
result = compare_weight_hypothesis(hypothesis)

if result == True:
    print("The average weight of male college students is greater than the average weight of female college students.")
elif result == False:
    print("The average weight of male college students is not greater than the average weight of female college students.")
else:
    print(result)
```

# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

```python
import scipy.stats as stats

def two_sample_hypothesis_test(sample1, sample2, alpha, alternative):
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, alternative=alternative)

    if p_value < alpha:
        return "Reject the null hypothesis"
    else:
        return "Fail to reject the null hypothesis"

# Example data for two samples
sample1 = [65, 72, 68, 60, 70, 75, 68, 71, 67, 73]  # Example weights of males
sample2 = [58, 65, 62, 55, 61, 59, 63, 57, 61, 60]  # Example weights of females

# Set the significance level (alpha) and alternative hypothesis
alpha = 0.05  # Example significance level
alternative = "two-sided"  # "two-sided", "greater", or "less"

# Perform the hypothesis test
result = two_sample_hypothesis_test(sample1, sample2, alpha, alternative)
print(result)
```

# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.


```python
import scipy.stats as stats

def two_sample_t_test(sample1, sample2, alpha):
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    if p_value < alpha:
        return "Reject the null hypothesis"
    else:
        return "Fail to reject the null hypothesis"

alpha = 0.05
sample1 = [65, 72, 68, 60, 70, 75, 68, 71, 67, 73]  # Example weights of males
sample2 = [58, 65, 62, 55, 61, 59, 63, 57, 61, 60]  # Example weights of females
result = two_sample_t_test(sample1, sample2, alpha)
print(result)
```

# Q11: What is Students t distribution? When to use the t-Distribution.
**Student's t-Distribution** is a probability distribution that is used to make inferences about population means when the sample size is small or when the population standard deviation is unknown.

Use the t-Distribution when:

- Population standard deviation is unknown.

# Q12: What is t-statistic? State the formula for t-statistic.
**t-statistic** is a value used in hypothesis testing to determine if there's a significant difference between sample means.

Formula: t = (sample mean - population mean) / (sample standard deviation / sample size)

# Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be 500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.
Using the formula for the confidence interval:
Confidence Interval = Sample Mean (Critical Value) * (Sample Standard Deviation / Sample Size)

With a 95% confidence level, the critical value for a two-tailed test is approximately 1.96.

Confidence Interval = $500  (1.96) * ($50 / 50)
Confidence Interval  $500  $13.94

Confidence Interval  [$486.06, $513.94]


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given values
sample_mean = 500  # Sample mean revenue
sample_std_dev = 50  # Sample standard deviation
sample_size = 50  # Sample size
confidence_level = 0.95

# Calculate the critical value from the t-distribution
degrees_of_freedom = sample_size - 1
critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

# Generate x values for plotting
x = np.linspace(sample_mean - 3*sample_std_dev, sample_mean + 3*sample_std_dev, 500)

# Generate normal distribution data for plotting
normal_dist = stats.norm.pdf(x, loc=sample_mean, scale=sample_std_dev)

# Create the plot
plt.plot(x, normal_dist, label='Population Distribution')
plt.axvline(x=sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(x=confidence_interval_lower, color='blue', linestyle='--', label='Confidence Interval Lower')
plt.axvline(x=confidence_interval_upper, color='blue', linestyle='--', label='Confidence Interval Upper')
plt.xlabel('Revenue')
plt.ylabel('Probability Density')
plt.title("Population Mean Revenue Estimation")
plt.legend()
plt.show()

print(f"Estimated population mean revenue: {sample_mean}")
print(f"Confidence Interval: [{confidence_interval_lower:.2f}, {confidence_interval_upper:.2f}]")
```

# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
Null Hypothesis (H0): The new drug does not decrease blood pressure by 10 mmHg.
Alternative Hypothesis (Ha): The new drug decreases blood pressure by 10 mmHg.

Using the formula for the t-statistic: t = (sample mean - hypothesized mean) / (sample standard deviation / sample size)

t = (8 - 10) / (3 / 100) = -2 / 0.3  -6.67

Using a t-distribution table or calculator, find the p-value associated with t = -6.67. Since the p-value is very small (much less than 0.05), we reject the null hypothesis. There is evidence to suggest that the new drug

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given values
sample_mean = 8  # Sample mean decrease in blood pressure
sample_std_dev = 3  # Sample standard deviation
sample_size = 100  # Sample size
population_mean = 10  # Hypothesized population mean decrease in blood pressure
significance_level = 0.05

# Generate t-distribution data
x = np.linspace(-5, 5, 500)
t_dist = stats.t.pdf(x, df=sample_size - 1)

# Create the t-distribution plot
plt.plot(x, t_dist, label='t-distribution')
plt.axvline(x=sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(x=population_mean, color='blue', linestyle='--', label='Hypothesized Mean')
plt.xlabel('t')
plt.ylabel('Probability Density')
plt.title("Student's t-Distribution")
plt.legend()
plt.show()

# Perform t-test
t_statistic, p_value = stats.ttest_1samp([sample_mean], popmean=population_mean)

# Print the results of the t-test
if p_value < significance_level:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
