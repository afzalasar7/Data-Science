# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

**Answer:**

To calculate the confidence interval for a sample mean, you can use the formula:

Confidence Interval = Sample Mean ± (Critical Value) * (Sample Standard Deviation / √Sample Size)

Given:
- Sample Mean (x̄) = 50
- Sample Standard Deviation (s) = 5
- Sample Size (n) = ?
- Confidence Level = 95%

For a 95% confidence level and assuming a large enough sample size (usually n > 30), the critical z-value is approximately ±1.96 (look up in standard normal distribution table).

Calculating the confidence interval:
Confidence Interval = 50 ± 1.96 * (5 / √n)

Interpretation: This means we are 95% confident that the true population mean lies within the calculated interval.

# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

**Answer:**

Interpretation: We are 95% confident that the true proportion of individuals who smoke in the population lies within the calculated interval.


```python
import statsmodels.stats.proportion as proportion

sample_size = 500
sample_proportion = 60 / sample_size

conf_interval = proportion.proportion_confint(count=60, nobs=sample_size, alpha=0.05, method='normal')

print("Confidence Interval:", conf_interval)
```

# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

**Answer:**
Interpretation: We are 90% confident that the true population mean lies within the calculated interval.

```python
import scipy.stats as stats

sample_mean = 75
sample_std_dev = 12
sample_size = 1

alpha = 0.10  # 90% confidence level
critical_value = stats.norm.ppf(1 - alpha/2)

margin_of_error = critical_value * (sample_std_dev / (sample_size ** 0.5))

confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print("Confidence Interval:", (confidence_interval_lower, confidence_interval_upper))
```

# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

degrees_of_freedom = 10
x = np.linspace(0, 30, 500)  # Range of chi-square values

chi2_distribution = stats.chi2.pdf(x, df=degrees_of_freedom)

plt.plot(x, chi2_distribution, label=f"df = {degrees_of_freedom}")
plt.fill_between(x, chi2_distribution, where=(x >= 15), alpha=0.5, color='blue', label="Area = P(X >= 15)")

plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution')
plt.legend()
plt.show()
```

# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

**Answer:**
Interpretation: If the p-value is less than the significance level (0.05), we reject the null hypothesis and conclude that the observed frequencies do not match the expected frequencies of a fair coin.


```python
observed_tails = 45
expected_tails = 0.5 * 100  # Expected number of tails in a fair coin flip

observed_frequencies = [observed_tails, 100 - observed_tails]  # Observed frequencies
expected_frequencies = [expected_tails, 100 - expected_tails]  # Expected frequencies

chi2_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

alpha = 0.05
if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

print(result)
```

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

**Answer:**
Interpretation: If the calculated t-statistic falls outside the critical value range, we reject the null hypothesis and conclude that the population mean is significantly different from 70.

```python
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70

t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

alpha = 0.05
critical_value = stats.t.ppf(1 - alpha/2, df=sample_size - 1)

if t_statistic > critical_value or t_statistic < -critical_value:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

print(result)
```
